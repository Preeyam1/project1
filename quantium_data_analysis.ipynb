# Outlier Region Analysis

```python
Q1 = data['TOT_SALES'].quantile(0.25)
Q3 = data['TOT_SALES'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

outliers = data[(data['TOT_SALES'] < lower_bound) | (data['TOT_SALES'] > upper_bound)]
outliers
```

```python
premium_customer_sales = outliers.groupby('PREMIUM_CUSTOMER')['TOT_SALES'].sum()

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(premium_customer_sales, labels=premium_customer_sales.index, autopct='%1.1f%%')
plt.title('Total Sales by Premium Customer')
plt.axis('equal')

# Display the pie chart
plt.show()

#insights

'''From the analysis of the pie chart, it is evident that the mainstream category customers contribute the highest percentage of total sales, accounting for 38.4%. 
This indicates that targeting and catering to the needs of mainstream customers can have a significant impact on overall sales. Additionally, the budget category 
customers also make a substantial contribution to total sales. Understanding the preferences and purchasing behavior of these customer segments can help businesses 
optimize their marketing strategies and product offerings to maximize sales and revenue.'''
```

```python
sns.swarmplot(x='PROD_QTY', y='TOT_SALES', hue='PREMIUM_CUSTOMER', data=outliers)
plt.xlabel('Product Quantity')
plt.ylabel('Total Sales')
plt.title('Analysis of Outliers: Product Quantity vs. Total Sales by Premium Customer')
plt.show()

#insight

'''The analysis of the outliers reveals that product quantity 5 has the highest number of sales, followed by product quantity 4. This suggests that customers tend
to purchase larger quantities of products when they fall into the outlier range of total sales. This insight can be valuable for the business as it indicates the 
potential for increased revenue by promoting and targeting products with higher quantities to customers who show a preference for purchasing in larger quantities.'''
```

```python
sales_by_category = outliers.groupby(['PREMIUM_CUSTOMER', 'LIFESTAGE'])['TOT_SALES'].sum()

# Plot the bar chart
sales_by_category.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Premium Customer & Lifestage')
plt.ylabel('Total Sales')
plt.title('Total Sales for Outliers by Premium Customer & Lifestage')
plt.xticks(rotation=45)
plt.show()
```
